#Problem 1 Car Parking System

```python
def parking_system(capacity, logs):
    current = 0

    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT":
            if current > 0:
                current -= 1
    if current > capacity:
        status = "Alert: Parking capacity exceeded"
    elif current == capacity:
        status = "Full"
    else:
        status = "Available"
    return current, status
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

current, status = parking_system(capacity, logs)

print("Currently Parked Vehicles:", current)
print("Parking Status:", status)
```

#Problem 2 : Estimate Delivery Time

```python
def estimate_delivery_time(distance, traffic, weather):

    base_time = distance * 5
    total_time = base_time
    traffic = traffic.lower()

    if traffic == "medium":
        total_time += 5
    elif traffic == "high":
        total_time += 10
    weather = weather.lower()

    if weather == "rainy":
        total_time += 5
    elif weather == "stormy":
        total_time += 10

    return total_time
distance = int(input("Enter Distance in km :"))
traffic = input("Traffic:")
weather = input("Weather:")
eta = estimate_delivery_time(distance, traffic, weather)

print("Estimated Delivery Time:", eta, "minutes")
```

#Problem 3 Movie Theatre Seat Occupancy Analyzer


```python
def seat_occupancy_analyzer(total_seats, booked_seats):

    booked_count = len(booked_seats)

    occupancy = (booked_count / total_seats) * 100
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Available"
    if occupancy == 100:
        suggestion = "Open additional shows"
    else:
        suggestion = "No additional show needed"
    return occupancy, status, suggestion
total_seats = 200
booked_seats = [1] * 150   # creates list with 150 bookings

occupancy, status, suggestion = seat_occupancy_analyzer(total_seats, booked_seats)

print("Occupancy:", occupancy, "%")
print("Show Status:", status)
print("Suggestion:", suggestion)
```

# Problem 4: Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    if not cpu_readings:
        return 0, "No readings available"

    average_cpu = sum(cpu_readings) / len(cpu_readings)

    if average_cpu < 50:
        status = "Normal"
    elif 50 <= average_cpu < 80:
        status = "Warning"
    else:
        status = "Critical"

    return average_cpu, status

# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

average_cpu_load, server_status = classify_server_load(cpu_readings)

print(f"Average CPU Load: {average_cpu_load:.0f}%")
print(f"Server Status: {server_status}")
```

# Problem 5: Smart Classroom Resource Usage Monitor

```python
def resource_usage_monitor(usage_data):
    overuse_thresholds = {
        "Projector": 5, # Example threshold: 5 hours
        "AC": 8,        # Example threshold: 8 hours
        "Lights": 6     # Example threshold: 6 hours
    }

    overused_resources = []
    energy_alert = "No"

    for resource, hours in usage_data.items():
        if resource in overuse_thresholds and hours > overuse_thresholds[resource]:
            overused_resources.append(resource)
            energy_alert = "Yes"

    return overused_resources, energy_alert

# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

overused, alert = resource_usage_monitor(resource_usage)

print(f"Overused Resources: {', '.join(overused) if overused else 'None'}")
print(f"Energy Alert: {alert}")
```

# Problem Statement 6: Online Event Registration Capacity Controller

```python
def event_registration_controller(capacity, registrations):
    if registrations <= capacity:
        confirmed_registrations = registrations
        waitlisted_users = 0
        registration_status = "Open"
    else:
        confirmed_registrations = capacity
        waitlisted_users = registrations - capacity
        registration_status = "Closed"

    return confirmed_registrations, waitlisted_users, registration_status

# Sample Input
event_capacity = 100
registrations_count = 105

confirmed, waitlisted, status = event_registration_controller(event_capacity, registrations_count)

print(f"Confirmed Registrations: {confirmed}")
print(f"Waitlisted Users: {waitlisted}")
print(f"Registration Status: {status}")
```

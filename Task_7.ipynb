# Problem 1: Smart Light Controller

```python
class SmartLightController:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"
        print(f"{self.light_name} created with initial status: {self.status}")

    def turn_on(self):
        self.status = "ON"
        print(f"{self.light_name} is now ON.")

    def turn_off(self):
        self.status = "OFF"
        print(f"{self.light_name} is now OFF.")

    def display_status(self):
        print(f"{self.light_name} is {self.status}")
House = SmartLightController("Balcony")
House.turn_on()
House.display_status()
House.turn_off()
House.display_status()
```

#Problem 2: Employee ID Card System

```python
class EmployeeIDCardSystem:
    def __init__(self, name, employee_id, department):
        self.name = name
        self.employee_id = employee_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.employee_id}")
        print(f"Department: {self.department}")
employee1 = EmployeeIDCardSystem("Jeshu", "EMP102", "DataScience")
employee1.display_id_card()
```

#Problem 3: Mobile Contact Record

```python
class MobileContactRecord:
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number
        print("Contact Saved")

    def display_contact(self):
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_number}")

contact1 = MobileContactRecord("Jeshu", "9886557741")
contact1.display_contact()
```

#Problem 4: Product Price Tag Generator

```python
class ProductPriceTagGenerator:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    def generate_price_tag(self):
        print(f"Product: {self.product_name}")
        print(f"Price: ₹{self.price}")

# Sample Input:
product1 = ProductPriceTagGenerator("IPad", 21499)
product1.generate_price_tag()
```

#Problem 5: Movie Rating Display System

```python
class MovieRatingDisplaySystem:
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating

    def display_movie_details(self):
        print(f"Movie: {self.movie_name}")
        print(f"Rating: {self.rating} / 5")

# Sample Input:
movie1 = MovieRatingDisplaySystem("Varanasi", 4.8)
movie1.display_movie_details()
```

#Problem 6: Delivery Address Manager

```python
class DeliveryAddressManager:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

# Sample Input:
customer1 = DeliveryAddressManager("Jeshu", "Hyderabad")
customer1.display_delivery_details()
```

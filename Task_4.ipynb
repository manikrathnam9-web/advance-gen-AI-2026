```python
valid_plans = {199, 299, 399, 599}
def recharge(plan):
  if plan<50:
    return False
  if plan in valid_plans:
    return True
  else:
    return False
while True:
  plan=input("Enter the plan: ")
  if plan.isdigit():
    plan=int(plan)
    if recharge(plan)==True:
      print("Recharge Successful")
      break
    else:
      print("Recharge Unsuccessful. Try again")
  else:
    print("Invalid. Please enter in number only")
```

#Problem3

```python
def calculate_and_print_result(marks):
  if not marks:
    print("No marks provided. Cannot calculate average.")
    return

  average = sum(marks) / len(marks)
  print(f"Average marks: {average:.2f}")

  if average >= 50:
    print("Pass")
  else:
    print("Fail")

# Example usage:
student_marks1 = [70, 85, 60, 92, 45]
print("\nStudent 1:")
calculate_and_print_result(student_marks1)

student_marks2 = [30, 40, 48, 55, 35]
print("\nStudent 2:")
calculate_and_print_result(student_marks2)

student_marks3 = []
print("\nStudent 3 (empty list):")
calculate_and_print_result(student_marks3)
```

#Problem4

```python
def calculate_cab_fare(distance_km, is_peak_hour):
  base_fare = 50
  per_km_charge = 12

  fare = base_fare + (distance_km * per_km_charge)

  if is_peak_hour:
    fare *= 1.25 # 25% extra for peak hour

  return fare

while True:
  try:
    distance_input = input("Enter distance traveled in km: ")
    distance_km = float(distance_input)
    if distance_km <= 0:
      print("Distance must be a positive number. Please try again.")
      continue

    peak_hour_input = input("Is it a peak hour? (yes/no): ").lower()
    if peak_hour_input not in ['yes', 'no']:
      print("Please enter 'yes' or 'no' for peak hour. Try again.")
      continue
    is_peak_hour = (peak_hour_input == 'yes')

    total_fare = calculate_cab_fare(distance_km, is_peak_hour)
    print(f"The total cab fare is: ₹{total_fare:.2f}")
    break # Exit loop if inputs are valid

  except ValueError:
    print("Invalid input. Please enter a number for distance. Try again.")
```

#Problem5

```python
def check_eligibility(attendance_list):
  if not attendance_list:
    return "Attendance list is empty. Cannot determine eligibility."

  present_count = attendance_list.count('P')
  total_classes = len(attendance_list)

  attendance_percentage = (present_count / total_classes) * 100

  print(f"Attendance: {present_count}/{total_classes} ({attendance_percentage:.2f}%)")

  if attendance_percentage >= 75:
    return "Eligible"
  else:
    return "Not Eligible"

# Example usage:
student_attendance1 = ['P', 'P', 'A', 'P', 'P', 'P', 'A', 'P', 'P', 'P'] # 80% attendance
print(f"\nStudent 1: {check_eligibility(student_attendance1)}")

student_attendance2 = ['P', 'P', 'A', 'P', 'A', 'P', 'A', 'P', 'A', 'P'] # 60% attendance
print(f"\nStudent 2: {check_eligibility(student_attendance2)}")

student_attendance3 = ['P', 'P', 'P', 'P', 'P'] # 100% attendance
print(f"\nStudent 3: {check_eligibility(student_attendance3)}")

student_attendance4 = [] # Empty list
print(f"\nStudent 4: {check_eligibility(student_attendance4)}")
```

```python
import re

def check_password_strength(password):
  # Rule 1: Minimum length of 8
  if len(password) < 8:
    return False, "Password must be at least 8 characters long."

  # Rule 2: Must contain at least one digit
  if not re.search(r"\d", password):
    return False, "Password must contain at least one digit."

  # Rule 3: Must contain at least one special character (@#$)
  if not re.search(r"[@#$]", password):
    return False, "Password must contain at least one special character (@#$)."

  return True, "Password is strong."

# Example usage:
print("\n--- Password Strength Check Examples ---")

pass1 = "MySecurePwd1@"
is_strong, message = check_password_strength(pass1)
print(f"Password: '{pass1}' -> Strong: {is_strong}, Message: {message}")

pass2 = "short1@"
is_strong, message = check_password_strength(pass2)
print(f"Password: '{pass2}' -> Strong: {is_strong}, Message: {message}")

pass3 = "NoDigits!@#"
is_strong, message = check_password_strength(pass3)
print(f"Password: '{pass3}' -> Strong: {is_strong}, Message: {message}")

pass4 = "NoSpecialChar1"
is_strong, message = check_password_strength(pass4)
print(f"Password: '{pass4}' -> Strong: {is_strong}, Message: {message}")

pass5 = "ValidPassword1#"
is_strong, message = check_password_strength(pass5)
print(f"Password: '{pass5}' -> Strong: {is_strong}, Message: {message}")
```

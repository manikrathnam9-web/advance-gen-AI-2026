#Problem1

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Finding the highest score
highest_score = max(employees.values())

# Finding all employees with the highest score
top_performers = [name for name, score in employees.items() if score == highest_score]

output = f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})"

print(output)
```

#Problem2

```python
import collections
import string

search_query = "Buy mobile phone buy phone online"

search_query_lower = search_query.lower()

translator = str.maketrans('', '', string.punctuation)
search_query_no_punct = search_query_lower.translate(translator)

keywords = search_query_no_punct.split()

keyword_counts = collections.Counter(keywords)

filtered_keywords = {word: count for word, count in keyword_counts.items() if count > 1}

print(filtered_keywords)
```

#Problem3

```python
import collections

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domains = [email.split('@')[1] for email in emails]

domain_counts = collections.Counter(domains)

total_emails = len(emails)

for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

#Problem4

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []
for hour_index, reading_value in enumerate(sensor_readings):
    if reading_value % 2 == 0:
        valid_readings.append((hour_index, reading_value))

print(f"Valid Sensor Readings (Hour, Value):\n{valid_readings}")
```

#Problem5

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

average_sales = sum(sales) / len(sales)

print(f"Average Daily Sales: {average_sales:.2f}")

print("Sales Spikes Detected:")
for day_index, sale_value in enumerate(sales):
    if sale_value > average_sales * 1.30:

        print(f"Day {day_index + 1}: {sale_value}")
```

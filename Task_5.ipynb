#Problem1

```python
def calculate_total_likes(likes):
  total = 0
  for i in likes:
    total += i
  return total

def get_post_status(total_likes_count):
  if total_likes_count >= 1000:
    return "Viral post"
  else:
    return "Normal post"

Jeshlikes = [200, 300, 400, 150]
total_likes_sum = calculate_total_likes(Jeshlikes)
status = get_post_status(total_likes_sum)

print("Total likes: ", total_likes_sum)
print("Post status:", status)
```

#Problem3

```python
def analyze_rainfall(rainfall_data, required_level):
  total_rainfall = 0
  for rain in rainfall_data:
    total_rainfall += rain

  if not rainfall_data: # Handle case where rainfall_data is empty to avoid division by zero
    average_rainfall = 0
  else:
    average_rainfall = total_rainfall / len(rainfall_data)

  if average_rainfall >= required_level:
    status = "Adequate Rainfall"
  else:
    status = "Inadequate Rainfall"

  print(f"Average Rainfall: {average_rainfall:.0f}")
  print(f"Rainfall Status: {status}")

# Example Usage:
seasonal_rainfall = [60, 80, 75, 70, 90]
required_rainfall_level = 70
analyze_rainfall(seasonal_rainfall, required_rainfall_level)
```

#Problem4

```python
def detect_duplicate_usernames(usernames):
  seen_usernames = set()
  for user in usernames:
    if user in seen_usernames:
      return "Yes" # Duplicates found
    seen_usernames.add(user)
  return "No" # No duplicates found

# Example Usage:
user_list1 = ["alice", "bob", "charlie", "alice"]
duplicates1 = detect_duplicate_usernames(user_list1)
print(f"Duplicate Accounts Found: {duplicates1}")

user_list2 = ["david", "eve", "frank"]
duplicates2 = detect_duplicate_usernames(user_list2)
print(f"Duplicate Accounts Found: {duplicates2}")
```

#Problem6

```python
def filter_premium_crops(prices):
  premium_crops = []
  for price in prices:
    if price > 2000:
      premium_crops.append(price)
  return premium_crops

# Example Usage:
crop_prices = [1800, 2500, 1500, 3200, 1900]
premium_list = filter_premium_crops(crop_prices)
print(f"Premium Crops: {premium_list}")
```

```python
def evaluate_application_health(error_count):
  if error_count == 0:
    return "Healthy"
  elif error_count <= 5:
    return "Minor Issues"
  else:
    return "Critical Issues"

# Example Usage:
errors1 = 0
status1 = evaluate_application_health(errors1)
print(f"Error Count: {errors1}\nSystem Status: {status1}")

errors2 = 3
status2 = evaluate_application_health(errors2)
print(f"\nError Count: {errors2}\nSystem Status: {status2}")

errors3 = 7
status3 = evaluate_application_health(errors3)
print(f"\nError Count: {errors3}\nSystem Status: {status3}")
```

```python
def validate_transaction(amount):
  daily_limit = 50000
  if amount <= daily_limit:
    return "Approved"
  else:
    return "Rejected"

# Example Usage:
transaction_amount_1 = 45000
transaction_status_1 = validate_transaction(transaction_amount_1)
print(f"Transaction Amount: {transaction_amount_1}")
print(f"Transaction Status: {transaction_status_1}")

transaction_amount_2 = 60000
transaction_status_2 = validate_transaction(transaction_amount_2)
print(f"\nTransaction Amount: {transaction_amount_2}")
print(f"Transaction Status: {transaction_status_2}")
```

#Problem9

```python
def check_exam_eligibility(attended_classes, total_classes):
  if total_classes == 0:
    attendance_percentage = 0.0
  else:
    attendance_percentage = (attended_classes / total_classes) * 100

  if attendance_percentage >= 75:
    eligibility_status = "Eligible"
  else:
    eligibility_status = "Not Eligible"

  print(f"Attendance Percentage: {attendance_percentage:.1f}")
  print(f"Exam Eligibility: {eligibility_status}")

# Example Usage:
# Scenario 1: Eligible student
attended_1 = 80
total_1 = 100
check_exam_eligibility(attended_1, total_1)

print("\n") # Add a newline for better readability between examples

# Scenario 2: Not Eligible student
attended_2 = 60
total_2 = 100
check_exam_eligibility(attended_2, total_2)
```

```python
def calculate_electricity_bill(units_consumed):
  total_bill = 0

  if units_consumed <= 100:
    total_bill = units_consumed * 3
  elif units_consumed <= 200:
    total_bill = (100 * 3) + ((units_consumed - 100) * 5)
  else:
    total_bill = (100 * 3) + (100 * 5) + ((units_consumed - 200) * 7)

  usage_level = ""
  if total_bill < 500:
    usage_level = "Low Usage"
  elif 500 <= total_bill <= 1500:
    usage_level = "Moderate Usage"
  else:
    usage_level = "High Usage"

  return total_bill, usage_level

# Example Usage:
units1 = 80
bill1, usage1 = calculate_electricity_bill(units1)
print(f"Units Consumed: {units1}, Total Bill: ₹{bill1}, Usage Level: {usage1}")

units2 = 150
bill2, usage2 = calculate_electricity_bill(units2)
print(f"Units Consumed: {units2}, Total Bill: ₹{bill2}, Usage Level: {usage2}")

units3 = 250
bill3, usage3 = calculate_electricity_bill(units3)
print(f"Units Consumed: {units3}, Total Bill: ₹{bill3}, Usage Level: {usage3}")

units4 = 300
bill4, usage4 = calculate_electricity_bill(units4)
print(f"Units Consumed: {units4}, Total Bill: ₹{bill4}, Usage Level: {usage4}")
```

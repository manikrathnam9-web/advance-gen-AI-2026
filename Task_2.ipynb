#Problem4

```python
prices = [450, 1200, 899, 1500, 300]
count_above_1000 = 0

for price in prices:
    if price > 1000:
        count_above_1000 += 1

print(f"Products above 1000: {count_above_1000}")
```

#Problem5

```python
attendance = ["P", "P", "A", "P", "P"]
total_days = len(attendance)
present_days = attendance.count("P")

if total_days > 0:
    attendance_percentage = (present_days / total_days) * 100
else:
    attendance_percentage = 0.0

print(f"Attendance Percentage: {attendance_percentage}")
```

#Problem10

```python
marks = [70, 80, 65, 75]
total_marks = sum(marks)
number_of_subjects = len(marks)

if number_of_subjects > 0:
    average_marks = total_marks / number_of_subjects
else:
    average_marks = 0.0

print(f"Average Marks: {average_marks}")
```
